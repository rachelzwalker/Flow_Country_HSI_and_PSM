```python
import pandas as pd
import seaborn as sns
```

```python
ndvi = pd.read_csv("../../VI_vector_data/NDVI data.csv")
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=ndvi)

ax.set_xlabel("Site", fontsize = 12)
ax.set_ylabel("NDVI", fontsize = 12)

# NB: Proportion of the IQR past the low and high quartiles to extend the plot whiskers. 
# Points outside this range will be identified as outliers.
```

ndvi analysis:
Ragosti - 0.79-0.84 more likely to be normal conditions, 0.74-0.80 more likely to be warmer, 0.72-0.79 more likely to be warmer with reduced precipitation, 0.70-0.80 more likely to have reduced precipitation
Rastogi 2018 - surprised that low NDVI correlated with areas with more Sphagnum
Pajula and Purre 2021 - low = Sphagnum an sparse vascular vegetation, high = sedges

```python
pri = pd.read_csv("../../VI_vector_data/pri data.csv")
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=pri)

ax.set_xlabel("Site", fontsize = 12)
ax.set_ylabel("PRI", fontsize = 12)
```

pri analysis - Ragosti - -0.125--0.08 more likely to be normal conditions, -0.125--0.10 more likely to be warmer, -0.125--0.095 more likely to be warmer with reduced precipitation, -0.11--0.085 more likely to have reduced precipitation. Basically all in the normal/reduced precipitation ranges

PRI strongly correlated with cover type (shrubs and bare beat) Cole et al 2013

```python
msi = pd.read_csv("../../VI_vector_data/msi data.csv")
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=msi)

ax.set_xlabel("Site", fontsize = 12)
ax.set_ylabel("MSI", fontsize = 12)
```

MSI significantly correlated with volumetric moisture content and water-table depth (better correlations in the field) - Harris and Bryant 2009

MSI also showed strong correlations with near-surface wetness, although the range of moisture conditions over which teh index was able to detect change was highly dependent on Sphagnum species - Harris et al 2005

```python
ndwi = pd.read_csv("../../VI_vector_data/ndwi data.csv")
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=ndwi)

ax.set_xlabel("Site", fontsize = 12)
ax.set_ylabel("NDWI", fontsize = 12)
```
