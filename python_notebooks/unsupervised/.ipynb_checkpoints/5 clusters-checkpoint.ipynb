```python
import spectral as spec
import spectral.io.envi as envi
import numpy as np
import matplotlib
```

```python
cl = envi.open('../../preprocessed_data/dimensionality_reduction/bsq/cross_lochs_2.hdr', '../../preprocessed_data/dimensionality_reduction/bsq/cross_lochs_2.bsq')
view = spec.imshow(cl,bands=(55,27,20),stretch=0.05,title="RGB Image of Cross Lochs")
print(view)
```

```python
#Type help(kmeans) to show how the function works. To run the k-means algorithm on the image and create 5 clusters, 
# using a maximum of 50 iterations
(m,c) = spec.kmeans(cl,5,50)
print(c.shape)
```

```python
%matplotlib inline
import pylab
pylab.figure()
for i in range(c.shape[0]):
    pylab.plot(c[i])
pylab.show
pylab.title('Spectral Classes from K-Means Clustering for Cross Lochs')
pylab.xlabel('Bands (with Water Vapor Windows Removed)')
pylab.ylabel('Reflectance')
```

```python
#%matplotlib notebook
import matplotlib.pyplot as plt
from PIL import Image
view = spec.imshow(cl, bands=(55,27,20),stretch=0.01, classes=m) #NB will need to change RGB bands once determined which they are (as above)
view.set_display_mode('overlay')

view.class_alpha = 0.5 #set transparency
plt.axis('off')
plt.savefig("test.tif", bbox_inches='tight')

#view.show_data
```
