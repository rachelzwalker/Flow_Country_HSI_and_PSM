```python
import pandas as pd
import geopandas
cross_lochs = pd.read_csv('../cross_lochs_gdf_pft.csv')

# cross_lochs_focused = cross_lochs[cross_lochs.columns[360:364]]

cross_lochs.head()
```

```python
from sklearn.cluster import KMeans

# kmeans = KMeans(n_clusters=5, random_state=0).fit(cross_lochs_gdf[1:359])

kmeans = KMeans(n_clusters=5)

x = cross_lochs[cross_lochs.columns[1:359]]

y = kmeans.fit_predict(x)

y_df = pd. DataFrame(y, columns=['Cluster'])

cross_lochs['Cluster'] = y_df

print(cross_lochs.head())
```

```python
y_df.describe()
```

```python
cross_lochs['geometry'] = geopandas.GeoSeries.from_wkt(cross_lochs['geometry'])

cross_lochs_gdf = geopandas.GeoDataFrame(cross_lochs, geometry='geometry')

# cross_lochs_gdf.head(2)

# cross_lochs_focused = cross_lochs[cross_lochs_gdf.columns[360:365]]

cross_lochs_gdf.head()

# gdf.head()
```
