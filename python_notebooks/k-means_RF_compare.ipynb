```python
import pandas as pd
import geopandas

kmc_full = pd.read_csv('unsupervised/cross_lochs_kmc_alt.csv')
rf_test = pd.read_csv('supervised/cross_lochs_results_gdf.csv')
rf_train = pd.read_csv('supervised/cross_lochs_train_gdf.csv')
```

```python
kmc = kmc_full[["geometry", "Cluster"]]

kmc
```

```python
rf_both = [rf_train, rf_test]
rf = pd.concat(rf_both)
```

```python
import numpy as np

rf["cluster_approx"] = ""

# for index, row in kmc.iterrows():
#     if rf.iloc[index,3] == "water":
#         rf["cluster_approx"] = 1
#     else:
#         0

# rf

# for i in rf['PFT']:
#     if 'water' in i:
#         rf['cluster_approx'] = 1
#     if 'rushes_sedges' in i:
#         rf['cluster_approx'] = 2
#     if 'shrub_sphagnum' in i:
#         rf['cluster_approx'] = 3
#     if 'pool_bogbean' in i:
#         rf['cluster_approx'] = 4
#     else:
#         rf['cluster_approx'] = 0

rf['cluster_approx'] = pd.np.where(rf.PFT.str.contains("water"), 1,
                   pd.np.where(rf.PFT.str.contains("rushes"), 2,
                   pd.np.where(rf.PFT.str.contains("shrub"), 3, pd.np.where(rf.PFT.str.contains("pool"), 4, 0))))

rf
```

```python
rf["cluster_approx_pred"] = ""

rf['clf_pred'] = rf['clf_pred'].fillna("train")

rf['cluster_approx_pred'] = pd.np.where(rf.clf_pred.str.contains("water"), 1,
                   pd.np.where(rf.clf_pred.str.contains("rushes"), 2, pd.np.where(rf.clf_pred.str.contains("train"), 99,
                   pd.np.where(rf.clf_pred.str.contains("shrub"), 3, pd.np.where(rf.clf_pred.str.contains("pool"), 4, 0)))))

rf
```

```python
rf['cluster_rf'] = ""

rf['cluster_rf'] = pd.np.where(rf.cluster_approx_pred == 99, rf.cluster_approx, rf.cluster_approx_pred)

rf
```

```python
rf_kmc = pd.merge(rf, kmc)

rf_kmc
```

```python
y_cluster = rf_kmc["Cluster"]
y_rf = rf_kmc["cluster_rf"]

df_confusion = pd.crosstab(y_cluster, y_rf)
```

```python
df_confusion
```

```python
from sklearn.metrics import accuracy_score

accuracy_score = accuracy_score(y_cluster, y_rf)

accuracy_score
```

```python
del rf_kmc["clf_pred"]
del rf_kmc["cluster_approx"]
del rf_kmc["cluster_approx_pred"]

rf_kmc
```

```python
rf_kmc['geometry'] = geopandas.GeoSeries.from_wkt(rf_kmc['geometry'])

rf_kmc_gdf = geopandas.GeoDataFrame(rf_kmc, geometry='geometry')

rf_kmc_gdf.head(2)
```
