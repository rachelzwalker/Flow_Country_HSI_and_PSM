```python
import numpy as np

rf["cluster_approx"] = ""


rf['cluster_approx'] = pd.np.where(rf.clf_pred_full.str.contains("water"), 1,
                   pd.np.where(rf.clf_pred_full.str.contains("pool_bogbean"), 2, pd.np.where(rf.clf_pred_full.str.contains("rushes"), 5,
                   pd.np.where(rf.clf_pred_full.str.contains("sphagnum_r"), 3, pd.np.where(rf.clf_pred_full.str.contains("calluna"), 4, 0)))))

rf
```

```python
rf_kmc = pd.merge(rf, kmc)

rf_kmc
```

```python
y_cluster = rf_kmc["Cluster"]
y_rf = rf_kmc["cluster_approx"]

df_confusion = pd.crosstab(y_cluster, y_rf)
```

```python
df_confusion
```

```python
from sklearn.metrics import accuracy_score

accuracy_score = accuracy_score(y_cluster, y_rf)

accuracy_score
```
